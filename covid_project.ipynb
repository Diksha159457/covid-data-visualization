```python
import pandas as pd
import folium

# Load dataset from GitHub
url = "https://raw.githubusercontent.com/datasets/covid-19/main/data/countries-aggregated.csv"
df = pd.read_csv(url)

df.head()

latest = df[df["Date"] == df["Date"].max()]
coords = {
    "India":[20.59,78.96],
    "US":[37.09,-95.71],
    "Brazil":[-14.23,-51.92],
    "France":[46.22,2.21],
    "United Kingdom":[55.37,-3.43]
}

latest["latitude"] = latest["Country"].map(lambda x: coords.get(x,[None,None])[0])
latest["longitude"] = latest["Country"].map(lambda x: coords.get(x,[None,None])[1])
m = folium.Map(location=[20,0], zoom_start=2, tiles="OpenStreetMap")
for _,row in latest.iterrows():
    if pd.notna(row["latitude"]):
        folium.CircleMarker(
            location=[row["latitude"],row["longitude"]],
            radius=6,
            popup=f"{row['Country']} : {row['Confirmed']}",
            fill=True
        ).add_to(m)


m.save("covid_world_map.html")
```
